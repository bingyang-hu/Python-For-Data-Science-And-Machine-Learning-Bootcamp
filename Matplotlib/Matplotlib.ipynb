```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
# functional method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

multiple plots

```python
plt.subplot(1,2,1) # # of rows, # of columns, # of plots
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

Object Oriented Method:
Create a figure object
define axes position

```python
# Object Oriented Method
fig = plt.figure() # create a figure object

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
```

```python

fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()


## We can iterate axes!
for current_ax in axes:
    current_ax.plot(x,y)
    
axes[0].set_title('First Plot')
```

#### Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

#### Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label= 'X Cubed')

ax.legend(loc=0)
```

#### Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=0.5,alpha = 1,linestyle= '--'
        ,marker='o',markersize=10,markerfacecolor='green',
       markeredgewidth=3,markeredgecolor='purple')

# either linewidth or lw is ok
# either linestyle or ls is ok
```

#### Control over axis appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
